```python
# 叠加平均
evoked = epochs.average()
```

```python
# 绘制地形图:第三个参数应该是一个整数，表示生成的样本数量
times = np.linspace(-2.0, 0.1, 8)
evoked.plot_topomap(times=times, colorbar=True)
```

```python
# 绘制某一特定时刻的地形图
# 此例绘制0.8s处，取0.75-0.85s的均值
evoked.plot_topomap(times=0.8, average=0.1)
```

```python
#绘制平均所有电极后的ERP
mne.viz.plot_compare_evokeds(evokeds=evoked, combine='mean')
```

```python
#绘制枕叶电极的平均ERP
mne.viz.plot_compare_evokeds(evokeds=evoked, picks=['O1', 'Oz', 'O2'], combine='mean')
```

```python
import os
import glob
import mne

data_type = 'stim'
subjects = glob.glob(os.path.join(r'D:\naturehuman\output', 'eeg_fif', data_type, '*.fif'))
grand_avg = []
for subject in subjects:
    epo = mne.read_epochs(subject)
    if epo.get_data().shape[0]:
        grand_avg += [epo.average()]
grand_avg = mne.grand_average(grand_avg)
```

```python
grand_avg
```

```python
times = np.linspace(-0.2, 4.0, 8)
grand_avg.plot_topomap(times=times, colorbar=True)
```
